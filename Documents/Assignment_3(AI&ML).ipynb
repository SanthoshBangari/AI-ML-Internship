```python
#Task5-Build ML model with linear regression (Target column is price)

df=df[['price','sqft_lot']]
```

```python
df
```

```python
x=df['sqft_lot']
```

```python
y=df['price']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('sqft_lot')
plt.ylabel('price')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=23)
```

```python
x_train=np.array(x_train).reshape(-1,1)
```

```python
x_train
```

```python
x_test=np.array(x_test).reshape(-1,1)
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Ir=LinearRegression()
```

```python
Ir.fit(x_train,y_train)
```

```python
c=Ir.intercept_
c
```

```python
m=Ir.coef_
m
```

```python
y_pred_train=m*x_train+c
y_pred_train.flatten()
```

```python
y_pred_train1=Ir.predict(x_train)
y_pred_train1
```

```python
plt.scatter(x_train,y_train)
plt.scatter(x_train,y_pred_train1,color='red')
plt.xlabel('sqft_lot')
plt.ylabel('price')
```

```python
Ir.fit(x_test,y_test)
```

```python
y_pred_test=m*x_test+c
y_pred_test.flatten()
```

```python
y_pred_test1=Ir.predict(x_test)
y_pred_test1
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_train,y_pred_train1,color='red')
plt.xlabel('sqft_lot')
plt.ylabel('price')
```
